```
global l_lendo,e_esc,e_espe,l_espe, ecriada, lcriada,ee,el,varl,vare,nleituras,nescritas,l_encer,e_encer,var_atual,var_rep,cont,ence,encl,execucoes

nleituras = nescritas = ecriada = lcriada = erroe = errol = l_lendo = e_esc = e_espe = l_espe = ee = el = l_encer = e_encer = cont = ence = encl = 0
lrep = erep = eav = lav = 0
execucoes = 0

def num_thread_tipo (nthreadsl,nthreadse):
    
    global nthe #número de threads escritoras
    global nthl #número de threads leitoras
    global vesc #vetor para verificar o número de execuções de cada thread escritora
    global vlei #vetor para verificar o número de execuções de cada thread leitora
    global thcl #vetor pra armazenar as threads leitoras criadas
    global thce #vetor pra armazenar as threads escritoras criadas
    global varl
    global vare
    

    nthe = nthreadse #numero de threads escritoras passado pelo usuário
    nthl = nthreadsl #numero de threads leitoras passado pelo usuário

    varl = list(0 for i in range(nthreadsl)) 
    vare = list(0 for i in range(nthreadsl)) 
    
    
    thcl = list(-1 for i in range(nthreadsl)) 
    thce = list(-1 for i in range(nthreadse))
    vesc = list(0 for i in range(nthreadse)) #inicializando o vetor todo com 0 para contabilizar todas as execuções de cada thread escritora
    vlei= list(0 for i in range(nthreadsl)) #inicializando o vetor todo com 0 para contabilizar todas as execuções de cada thread leitora


#Função chamada para passar o parâmetro passado pelo usuário de quantas leituras e escritas deveriam ser feitas por cada thread
def num_exec_thread (nlei,nesc):
    global nl
    global ne
    global var_rep
    global cont
    global nthe #número de threads escritoras
    global nthl #número de threads leitoras

    
    nl = nlei #numero de leitura que deve ser feita por cada thread leitora
    ne = nesc #numero de escrita que deve ser feita por cada thread escritora

    var_rep = list(0 for i in range(nthl*nl + nthe*ne)) 
   


#Função chamada quando uma thread escritora foi criada
def esc_criada(i,var):   
    global ecriada
    global thce


    thce.append(i) 
    ecriada += 1 #numero de thread escritora aumenta, irá ser usado depois para verificar se foi criado o número de threads pedido
    print("Escritora %d criada com Sucesso" %i)

#Função chamada quando uma thread leitora foi criada
def lei_criada(i,var):    
    global lcriada 
    global thcl


    thcl.append(i) 
    lcriada += 1 #numero de thread leitora aumenta, irá ser usado depois para verificar se foi criado o número de threads pedido
    print("Leitora %d criada com Sucesso" %i)

#Função chamada quando uma thread escritora entra pra fila de escritoras
def e_entrou_fila(i,var):
    global l_lendo, e_esc, e_espe,erro
    
    #Verificando se ela tinha que entrar pra fila mesmo ou se poderia ser executada:

    if(e_esc == 1 or l_lendo > 0): #Ela só deve entrar pra fila se já houver uma escritora escrevendo ou uma mais leitora lendo
        print("Escritora %d entrou na fila" %i) 
        erro += 0 
        e_espe +=1 #Como ela deveria mesmo ter entrado, a fila de escritoras aumenta
    
    else:
        erro += 1
        print("Escritora %d deveria ter executado, pois não tinha leitora lendo nem escritora escrevendo" %i)   

#Função chamada quando uma thread leitora entra pra fila de leitoras
def l_entrou_fila(i,var):
    global erro,e_esc,l_lendo,e_espe,l_espe
    
    #Verificando se ela tinha que entrar pra fila mesmo ou se poderia ser executada:

    if(e_esc == 1 or e_espe > 0): #Ela só deve entrar pra fila se já houver uma escritora escrevendo ou uma ou mais escritora esperando
        print("Leitora %d entrou na fila" %i)
        l_espe+=1 #Como ela deveria mesmo ter entrado, a fila de leitoras aumenta
        erro += 0


    else:
        print("e_esc: %d e e_esp: %d" %(e_esc,e_esp))
        erro += 1
        print("Leitora %d deveria ter executado, pois não tinha escritora escrevendo e nem escritora na fila" %i) 


#Função chamada quando uma thread escritora recebe sinal para sair do wait
def e_recebeu_sinal(i,var):
    global erro, l_parou,l_lendo,e_esc,l_espe,e_espe
    global ee

    #Verificando se ela podia ter recebido o sinal:
    #ELa só pode receber o sinal se não houver leitora lendo, no caso de ser um sinal enviado por uma leitora
    #Ou se não tem leitoras na fila, no caso de ser um sinal enviado por uma escritora

    if (l_lendo==0 or l_espe == 0): 
        erro += 0
        print("Escritora %d recebeu sinal e saiu da fila" %i)
        

    else: 
        erro += 0
        print("Escritora %d recebeu sinal errado ou alguma thread passou sua frente" %i) 
        ee = 1

#Função chamada quando uma thread leitora recebe sinal para sair do wait
def l_recebeu_sinal(i,var):
    global erro, e_esc,l_esp,l_espe
    global el

    #Verificando se ela podia ter recebido o sinal:
    #Ela só pode receber o sinal se não houver escritora escrevendo

    if (e_esc==0):
        erro += 0
        print("Leitora %d recebeu sinal e saiu da fila" %i)

    else: 
        print("Leitora %d recebeu sinal errado ou alguma thread passou sua frente" %i)
        el = 1 


#Função chamada quando uma thread escritora recebe sinal para sair do wait, mas ao verificar se a condição permanece válida, vé que não e não sai da fila
def e_continuou_fila(i,var):
    global erro, e_esc,l_lendo
    global ee

    #Verificando se ela devia permanecer na fila:
    #Ela só devia permanecer, se uma escritora começou a escrever ou uma leitora começou a ler

    if(e_esc==1 or l_lendo == 1):
        print("Leitora %d continua na fila" %i)
        ee-=1

    #Se ee for igual a 0, alguma thread passou sua frente, ela verificou isso e continuou na fila, não gerando erro
    if(ee==0):
        erro += 0
    else:
        erro += 1



#Função chamada quando uma thread leitora recebe sinal para sair do wait, mas ao verificar se a condição permanece válida, vé que não e não sai da fila
def l_continuou_fila(i,var):
    global e_esc
    global erro, el
    
    #Verificando se ela devia permanecer na fila:
    #Ela só deve continuar na fila se tem uma escritora escrevendo

    if(e_esc==1):
        print("Leitora %d continua na fila" %i)
        el-=1
    if(el==0):
        erro += 0
    else:
        erro += 1

#Função chamada quando uma thread escritora sai da fila de escritoras pra executar a escrita
def e_saiu_fila(i,var):
    global e_espe

    #Como já foi verficado que ela podia sai no função de recebeu sinal, não testei de novo
    print("Leitora %d saiu da fila" %i)
    e_espe -=1  #Fila de escritoras deminui

#Função chamada quando uma thread leitora sai da fila de leitoras pra executar a leitura
def l_saiu_fila(i,var):
    global l_espe

    #Como já foi verficado que ela podia sai no função de recebeu sinal, não testei de novo
    print("Leitora %d saiu da fila" %i)
    l_espe -=1 #Fila de leitoras deminui
 

#Função chamada quando uma escritora escreve
def escritora_escreveu(id_esc,var):
    global erroe
    global errose, e_esc
    global vare
    global nescritas
    global var_atual
    global var_rep
    global cont
    global execucoes

    execucoes+=1
    
    
    var_rep[cont] = "escrita" #Escrevendo em um vetor a operação para depois ser analisado a alternância entre as threads
    cont+=1
   
    
    nescritas+=1

    #Verificando se ela podia estar escrevendo:
    #Ela só podia escrever se não tinha leitora lendo ou escritora escrevendo

    if l_lendo != 0:
        print("     A %dª escrita do programa feita pela escritora %d não deveria ter acontecido pois tinha leitora escrevendo" %(nescritas,id_esc))
        erroe += 1

    if e_esc > 1:
        print("     A %dª escrita do programa feita pela escritora %d não deveria ter acontecido pois tinha escritora escrevendo" %(nescritas,id_esc))
        erroe = 1

    if(erroe == 0): #As condições foram satisfeitas e ela podia escrever
        e_esc = 1 
        print("Escritora %d escreveu" %id_esc)
        


    vesc[id_esc] += 1 #Aumenta o número de escrita realizada pela thread escritora em questão


#Função chamada quando uma leitora lê
def leitora_leu(id_lei,var):
    global errol
    global errosl
    global l_lendo
    global erro
    global e_esc
    global varl
    global nleituras
    global var_atual
    global var_rep
    global cont
    global execucoes

    execucoes+=1    
    nleituras += 1

    var_rep[cont] = "leitura" #Escrevendo em um vetor a operação para depois ser analisado a alternância entre as threads
    cont+=1

    l_lendo +=1 #contador que indica quantas threads leitoras estão lendo aumenta

    #Verificando se ela podia estar escrevendo:
    #Ela só podia ler se não tinha escritora escrevendo

    if e_esc != 0:
        errosl.append("     A %dª leitura do programa feita pela leitora %d não deveria ter acontecido pois tinha escritora escrevendo" %(nleituras,id_lei))
        erro += 1
        errol += 1
         
    else: #As condições foram satisfeitas e ela podia ler
        print("Leitora %d leu %d" %(id_lei,var))

    vlei[id_lei] += 1 #Aumenta o número de leitura realizada pela thread leitora em questão

#Função chamada quando uma thread escritora termina de escrever
def termina_escrita(i,var):
    global e_esc, erro
    print("Escritora %d terminou escrita" %i)

    e_esc = 0 # o contador de thread escritora escrevendo vai pra 0 pois só tinha uma escrevendo de acordo com as condições

    if(varl[i] == var):
        erro += 0
    else:
        erro += 1


#Função chamada quando uma thread leitora termina de ler
def termina_leitura(i,var):
    global l_lendo
    global varl
    global erro

    print("Leitora %d terminou leitura" %i)
    l_lendo -= 1 # o contador de thread leitora lendo diminui

    if(varl[i] == var): #outra maneira de testar se a condição foi quebrada, que é testando se a variável mudou enquando a leitora lia
        erro += 0
    else:
        erro += 1

#Função chamada quando uma thread leitora manda sinal pra escritora após terminar de ler
def l_manda_sinal_l(i,var):
    global l_lendo,erro

    #Verificando se ela podia mandar sinal:
    #Ela só podia mandar sinal se não tivesse mais nenhuma leitora lendo
    
    if(l_lendo) == 0:
        erro += 0;
        print("Leitora %d mandou sinal pras escritoras - Estratégia para garantir a ausência de Inanição" %i) 

    else:
        erro += 1
        print("Leitora não deveria ter mandado sinal pra escritora, pois ainda tinha %d leitora(s) lendo" %l_lendo)

#Função chamada quando uma thread escritora manda sinal pras leitoras após terminar de escrever
def e_manda_sinal_l(i,var):

    global l_espe,e_esc,erro

    #Verificando se ela podia mandar sinal:
    #Ela só podia mandar sinal se tivesse uma ou mais leitoras na fila de leitoras

    if(l_espe) > 0:
        print("Escritora %d mandou sinal pras leitoras - Estratégia para garantir a ausência de Inanição" %i)
        erro += 0
    else:
        erro += 1
        print("Escritora não deveria ter mandado sinal pra leitora, pois não tinha leitoras na fila")

#Função chamada quando uma thread escritora manda sinal pras escritora após terminar de escrever
def e_manda_sinal_e(i,var):
    global l_espe,e_esc,erro

    #Verificando se ela podia mandar sinal:
    #Ela só podia mandar sinal se não tivesse  leitoras na fila de leitoras

    if(l_espe == 0):
        erro += 0
        print("Escritora %d mandou sinal pras escritoras" %i)
        

    else:
        erro += 1
        print("Escritora não deveria ter mandado sinal pra escritora, pois tinha %d leitora(s) na fila" %l_espe)



#Função implentada para verificar se as condições necessárias para cada leitura foram cumpridas, utilizando dados obtidos com as funções chamadas acima

def leituras():
    global errol
    global errosl

    print("\nProcesso: Leituras")

    #As condições foram cumpridas se a variável errol permaneceu 0 após todas chamadas da função leitora_leu
    if(not(errol)): 
        print("     Em todas as vezes que uma leitora leu, não tinha escritora escrevendo")
        print("SUCESSO")

    else:
        for i in range(len(errosl)):
            print(errosl[i])

        print("FALHA")

#Função implentada para verificar se as condições necessárias para cada escrita foram cumpridas, utilizando dados obtidos com as funções chamadas acima

def escritas():
    global erroe
    global errose
    global e_esc

    print("\nProcesso: Escritas")

    #As condições foram cumpridas se a variável erroe permaneceu 0 após todas chamadas da função escritora_escreveu
    if(not(erroe)):
        print("     Em todas as vezes que uma escritora escreveu, não tinha leitoras lendo nem outras escritoras escrevendo")
        print("SUCESSO")

    else:
        #for i in range(len(errose)):
            #print(errose[i])

        print("FALHA")

#Função implementada para verificar se foram criadas o número de threads leitora e escritora pedido pelo usuário, utilizando dados obtidos na chamadas de funções acima

def threads_criadas():
    global ecriada,lcriada,nthe,nthl
    erro = 0

    print("\nProcesso: Criação das Threads")

    #Foi criado o número de threads escritores pedido se a variável "ecriada" (incrementada todas vezes que a função "esc_criada" foi chamada) for igual ao número passado pelo usuário
    if (ecriada==nthe):
        print("     %d threads escritoras criadas como pedido" %nthe)

    else:
        erro = 1
        print("     Das %d pedidas, %d threads escritoras não foram criadas" %(nthe,nthe-ecriada))

    #Foi criado o número de threads leitoras pedido se a variável "lcriada" (incrementada todas vezes que a função "lei_criada" foi chamada) for igual ao número passado pelo usuário

    if (lcriada==nthl):
        print("     %d threads leitoras criadas como pedido" %nthl)
    else:
        erro = 1
        print("     Das %d pedidas, %d threads leitoras não foram criadas" %(nthl,nthl-lcriada,))

    if(erro):
        print("FALHA")
    else:
        print("SUCESSO")

#Função implementada para verificar se cada thread realizou o número de execuções pedido pelo usuário
def cada_thread():

    print("\nProcesso: Número de Execuções de cada Thread")
    
    global nthe
    global nthl
    global nl
    global ne
    global vesc
    global vlei

    erro = 0

    #Verificando se cada thread leitora executou o número pedido de leituras pelo usuário utilizando o vetor "vlei[i]" 
    #que foi incrementado toda vez em a função "leitora_leu" foi chamada pela leitora i 
    for i in range (nthl):
        if (vlei[i] != nl):
            print("     Problema: A thread leitora %d executou % vezes ao invés de %d como deveria ser feito" %(i,vlei[i],nlei))
            erro = 1

    if(not(erro)):
        print("     Cada thread leitora executou %d vezes como pedido" %nl)

    erro = 0

    #Verificando se cada thread escritora executou o número pedido de escritas pelo usuário utilizando o vetor "vesc[i]" 
    #que foi incrementado toda vez em a função "escritora_escreveu" foi chamada pela escritora i 
    for i in range (nthe):

        if (vesc[i] != ne):
            print("     Problema: A thread escritora %d executou % vezes ao invés de %d como deveria ser feito" %(i,vesc[i],nesc))
            erro = 1

    if(not(erro)):
        print("     Cada thread escritora executou %d vezes como pedido" %ne)

    if(erro):
        print("FALHA")
    else:
        print("SUCESSO")

#Funcção chamada quando uma thread leitora encerra suas execuções
def lei_encerrada(i,var):

    global l_encer,lcriada,vlei,nl,erro,encl,execucoes,nthl
    encl +=1

    if(vlei[i] != nl):
        print("Thread leitora %d não deveria ter sido encerrada, pois executou %d das %d leituras exigidas" %(i,vlei[i],nl))
        erro = 1
    else:
        erro = 0
        l_encer = lcriada - 1

    

    if(encl==nthl):
        print("A ultima thread leitora se encerrou na execução n°%d" %execucoes)


#Função chamada quando uma thread escritora encerra suas execuções
def esc_encerrada(i,var):
    global e_encer,ecriada,vesc,ne,erro,varr,var_rep,ence,execucoes,nthe
    e_encer = ecriada - 1
    ence +=1

    if(vesc[i] != ne):
        print("Thread leitora %d não deveria ter sido encerrada, pois executou %d das %d leituras exigidas" %(i,vesc[i],ne))
        erro = 1
    else:
        erro = 0
        l_encer = lcriada - 1

    if(ence==nthe):
        print("A ultima thread escritora se encerrou na execução n°%d" %execucoes)

#Função para verificar a alternância entre as threads para garantir que não houve inanição
def verificando_inanicao():

    global var_rep #vetor que contem a ordem de leituras e escritas ocorridas
    #print(var_rep)
    global execucoes,nthe,nthl,ne,nl
    conta = 0
    opatual = var_rep[0]
    repe = []

    print("\nProcesso: Alternância entre as threads leitoras e escritoras\n")
    
    for i in range (len(var_rep)):

        if (opatual == var_rep[i]):
            conta += 1
            

        else:            
            repe.append(conta)
            conta = 1            
            opatual = var_rep[i]   

        if (i==len(var_rep)-1):
                repe.append(conta)

    maior = max(repe, key=int)

    print(repe)
    
    print("\n     O vetor mostra a quantidade de vezes threads do mesmo tipo foram executadas em seguida")
    print("     Sendo assim, em %d execuções:" %execucoes)
    print("         O maior numero de execuções de uma thread do mesmo tipo em seguida foi %d" %maior)
    print("         E durante a execução do programa, threads do tipo escritora e do tipo leitora alternaram %d vezes" %(len(repe)))

    if(maior == repe[len(repe)-1]):
        print("         Sendo que o maior número ocorreu no final, explicado pelo fato de que todas as theads de um tipo já terem se encerrado")

    if(maior< (nthe*ne)*0.3 and maior< (nthl*nl)*0.3):
        print("SUCESSO")
    

```
